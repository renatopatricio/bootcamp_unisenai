## Tratamento dos outliers 

```python
# Selecionar apenas as colunas numéricas
numeric_cols = df.select_dtypes(include='number').columns

# Função para encontrar outliers usando o método do IQR
def find_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]

# Visualizar os outliers de cada coluna numérica
for col in numeric_cols:
    outliers = find_outliers(col)
    print(f"Outliers na coluna '{col}':")
    print(outliers)
    print('-' * 40)
```

```python
# Selecionar apenas as colunas numéricas
numeric_cols = df.select_dtypes(include='number').columns

# DataFrame para armazenar os outliers
df_outliers = pd.DataFrame()

# Função para encontrar outliers usando o método do IQR
def find_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] < lower_bound) | (df[col] > upper_bound)]

# Para cada coluna numérica, encontrar e armazenar os outliers
for col in numeric_cols:
    outliers = find_outliers(col)
    if not outliers.empty:
        outliers = outliers.copy()
        outliers['coluna_outlier'] = col
        df_outliers = pd.concat([df_outliers, outliers], ignore_index=True)

# Salvar os outliers em um arquivo CSV
if not df_outliers.empty:
    df_outliers.to_csv('outliers.csv', index=False)
else:
    print('Nenhum outlier encontrado.')
```
